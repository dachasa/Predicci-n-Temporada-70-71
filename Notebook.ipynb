# Aplicando One-Hot Encoding

Para a las variables categóricas les deberemos aplicar la técnica one-hot encoding: Esta técnica crea 16 nuevas columnas(número de equipos en la temporada 70) en los datos de manera que en dichas columnas habrá un 1 si la columna corresponde con el equipo y 0 en caso contrario. 

Generaremos columnas de tipo dummy en las columnas de Equipo1 y Equipo2

```python
Equipo1_dummy=pd.get_dummies(Temporada70['Equipo1'],prefix='EquipoLocal')
Equipo2_dummy=pd.get_dummies(Temporada70['Equipo2'],prefix='EquipoVisitante')
```

Unimos las columnas de tipo dummy a la información general del archivo

```python
Temporada70=pd.concat([Temporada70,Equipo1_dummy],axis=1)
Temporada70=pd.concat([Temporada70,Equipo2_dummy],axis=1)
```

```python
Temporada70
```

```python
Temporada70=Temporada70.drop(['Equipo1','Equipo2'], axis= 1)
```

```python
Temporada70
```

Ordenamos las nuevas columnas, para que las primeras salgan si un equipo juega de local o visitante, después los datos de las 5 jornadas anteriores, y por último el resultado del partido.

#Partición del dataset

```python
X = Temporada70.values[:,:-1]
Y = Temporada70.values[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
(trainData, testData, trainLabels, testLabels) = train_test_split(X, Y, test_size=0.2, random_state=84)
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
for k in range(1,26,2):
    modelKNN = KNeighborsClassifier(n_neighbors=k)
    modelKNN.fit(trainData, trainLabels)
    score = modelKNN.score(testData,testLabels)
    print("k=%d, precisión=%.2f%%" % (k, score * 100))
```

```python
modelKNN=KNeighborsClassifier(n_neighbors=23)
modelKNN.fit(trainData, trainLabels)
```

```python
PrediccionesKNN=modelKNN.predict(testData)                      
```

```python
Comparacion=pd.DataFrame({'Real':testLabels,'PrediccionesKNN':PrediccionesKNN})
Comparacion.head(40)
```

```python
Comparacion.transpose()
```

```python
from sklearn.metrics import accuracy_score
accuracyKNN=accuracy_score(testLabels,PrediccionesKNN)
accuracyKNN
```

La precisión es del 0.525, bastante buena.

```python
print("MATRIZ DE CONFUSIÓN USANDO KNN")
from sklearn.metrics import confusion_matrix
print(confusion_matrix(testLabels,PrediccionesKNN))
```
